```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('/Users/huiqiongwu/Desktop/data_science_project/association-rule-mining/Bakery.csv')
df.head(5)
```

```python
df = df.groupby(['TransactionNo', 'Items'])['Items'].count().reset_index(name='Quantity')
df
```

```python
pivot_df = pd.pivot(df, index='TransactionNo', columns='Items', values='Quantity').fillna(0)
pivot_df
```

```python
## Encode values
def encode_units(x):
    if x <= 0:
        return False
    if x >= 1:
        return True
bakery_df = pivot_df.applymap(encode_units)
bakery_df.head()
```

```python
## Generate frequent itemsets
frequent_itemsets = apriori(bakery_df, min_support = 0.03, use_colnames = True)
frequent_itemsets
```

```python
## Decide association rules
association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.2)
```

Analysis:
The popular itemsets with at least 3% support are {coffee, bread}, {coffee, cake}, {coffee, medialuna}, {coffee, pastry}, {coffee, sandwich}, {coffee, tea}. Moreover, based on the values of lift and confidence, the most popuar transactions are {coffee, cake}, {coffee, medialuna}, {coffee, pastry}, {coffee, sandwich}.
